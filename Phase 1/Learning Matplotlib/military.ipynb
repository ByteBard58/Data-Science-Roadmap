# Military Spending of US
I found a dataset on Kaggle that contains the military spending of the US from 1960 to 2020. I will use this dataset to practice some shit.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/Sakib/Downloads/MilitarySpending.csv")
data.tail(7)
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Year"],data["DefenseBudget"],"r.-",label="US Military Budget")
plt.xlabel("Year")
plt.ylabel("Spending (in Billion USD)")
plt.title("US Mlitary spending from 1960 to 2020",fontdict={"fontsize":18})
plt.show()
```

```python
bins = np.arange(100,1000,100)
plt.hist(data["DefenseBudget"],bins=bins)
plt.xlabel("Spending (in Billion USD)")
plt.ylabel("Various index")
plt.title("US Military Spending Histogram")
plt.show()
```

```python
old = data.query("Year < 1980")["DefenseBudget"].sum()
plt.figure(figsize=(5,5))
medium = data.query("Year >= 1980 and Year < 2000")["DefenseBudget"].sum()
modern = data.query("Year >= 2000 and Year < 2025")["DefenseBudget"].sum()
plt.pie([old,medium,modern],labels=["Before 1980","Between 1980 and 2000","After 2000"],explode=(0,0.1,0.1),autopct="%.2f")
plt.title("US Military spending",fontdict={"fontsize":19})
plt.show()
```
