#### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Plotting our first graph

```python
x = [0,1,2,3,4]
y=[0,3,6,9,12]

# Setting the size for the plot (figsize controls the physical size of the figure in inches, dpi = dots per inch)
plt.figure(figsize=(6,4),dpi=120) 

# Take a look at the keyword arguments
plt.plot([0,1,2,3,4,5],[0,2,5,2,8,18],label="Shoccho",color="Green",linewidth=2,marker=".",markersize=10,markeredgecolor="cyan",linestyle="--")

# Short-hand format: "[color(r,y,c,b,g)][marker(.,>,<)][linestyle(-,--)]"
plt.plot(x,y,"r>--",label="Sakib")

# Creates a title on top of the graph
plt.title("Our First Freakin' Graph",fontdict={"fontname":"Arial","fontsize":20})

# Creates the level of the X axis
plt.xlabel("X AXIS",fontdict={"fontname":"Arial","fontsize":14})

# Creates the level of the Y axis
plt.ylabel("Y AXIS",fontdict={"fontname":"Arial","fontsize":14})  

# Sets the scale of the axis
plt.xticks([0,1,2,3,4,5,6,6.6])             

# Sets the scale of the axis
plt.yticks([0,3,6,9,12,15,18,18.6])         

# Adds a legend
plt.legend()      

# Shows the plot
plt.show()
```

```python
# 2nd graph
plt.style.use("dark_background")
x2 = np.arange(-10,10.5,0.5)
plt.plot(x2,x2**2,"b.-",label="Quadratic Graph")

plt.legend()
plt.title("Graph of a quadratic function", fontdict={"fontsize":16})
plt.xlabel("X AXIS", fontdict={"fontsize":12})
plt.ylabel("Y AXIS", fontdict={"fontsize":12})
plt.show()
```

```python
# 3rd Graph
plt.style.use("dark_background")
plt.plot(x2,x2**3,"r.-",label="Cubic Graph")
plt.title("Graph of cubic function",fontdict={"fontsize":16})
plt.xlabel("X AXIS",fontdict={"fontsize":13})
plt.ylabel("Y AXIS",fontdict={"fontsize":13})

plt.legend()
plt.show()
```

```python
#4th Graph
plt.style.use("dark_background")   
x3 = np.arange(-10,10.5,0.5)

plt.plot(x3,1/x3,"c.-",label="Reciprocal Graph")
plt.title("Graph of a reciprocal function", fontdict={"fontsize":16})
plt.xlabel("X AXIS", fontdict={"fontsize":12})
plt.ylabel("Y AXIS", fontdict={"fontsize":12})
plt.legend()
plt.savefig("reciprocal.png")   # Save the graph
plt.show()
```

```python
# 5th graph
plt.plot(x2,np.sin(x2),"y.-",label="Sine Graph")
plt.title("Sine Wave", fontdict={"fontsize":16})
plt.xlabel("X AXIS", fontdict={"fontsize":12})
plt.ylabel("Y AXIS", fontdict={"fontsize":12})
plt.legend()
plt.show()
```

#### Bar Charts

```python
labels = ["A","B","C","D"]
values = [10,20,30,25]
# plt.bar(x axis, y axis)
bars = plt.bar(labels,values)

# Setting hatch manually for each columns
# bars[0].set_hatch("/")
# bars[1].set_hatch("-")

# Setting hatch with iteration
hatches = ["/","-","+","."]
for i,bar in enumerate(bars):
  bar.set_hatch(hatches[i])

plt.show()
```
