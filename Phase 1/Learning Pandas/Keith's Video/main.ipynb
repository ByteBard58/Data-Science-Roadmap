### .fillna()
Fills the null spots with a certain value

.interpolate() detects the relation between the data in a particular column and returns values in respect of that relation

```python
coffee.fillna(coffee.interpolate())
```

Fills the data with mean of a column

```python
coffee.fillna(coffee["Units Sold"].mean())
```

#### Important keyword arguments for .fillna()
value= → what to fill with (scalar, dict, Series)

method='ffill' → forward-fill (use previous value)

method='bfill' → backward-fill (use next value)

limit=n → max number of NaNs to fill

inplace=True → modify original DataFrame

### .dropna()
Drops (deletes) the rows (by default, can changed using axis=1) which have null values

```python
coffee.dropna(how="any",subset="Units Sold")
```

#### Important keyword arguments of .dropna()
axis=0 → drop rows (default)

axis=1 → drop columns

how='any' → drop if any NaN (default)

how='all' → drop if all values are NaN

subset=['col1', 'col2'] → only check specific columns

inplace=True → apply changes directly

### Using .value_counts()

### Using Groupby
**Gneral Syntax**

df.groupby('column_to_group')['column_to_aggregate'].agg_function()

**In case of custom agg function:**

df.groupby(['column_to_group']).agg({"column":"operation(mean/sum/etc)"})

```python
coffee.groupby(["Coffee Type"])["Price"].sum()
```

```python
coffee.groupby(["Coffee Type"]).agg({
  "Price":"sum",
  "Units Sold":"mean"
})
```

#### Common Aggregations
| Function   | Purpose                       |
| ---------- | ----------------------------- |
| `.sum()`   | Total of values               |
| `.mean()`  | Average                       |
| `.count()` | Number of entries             |
| `.max()`   | Highest value                 |
| `.min()`   | Lowest value                  |
| `.size()`  | Like count but returns Series |
| `.agg()`   | Custom multiple functions     |


### Pivot Tables 
It’s like an Excel pivot table, but in code:
we reshape our data to get a summary — grouped and aggregated, shown in table format.

**General Syntax**

something.pivot_table(columns=something, index=something,values=something)

```python
coffee.pivot_table(columns="Day",index="Coffee Type",values="Units Sold",aggfunc="sum")        
```

**NOTE:** There is a similar function named .pivot(), the problem with that one is that it doesn't support any aggfunc like .pivot_table()

```python
coffee.pivot_table(columns="Day",index="Coffee Type",values="Units Sold",aggfunc=["sum","mean"])
# Multiple aggfuncs are also valid to use        
```

#### .rolling()

The `.rolling()` function in pandas creates a moving window view over a Series or DataFrame, allowing you to perform calculations (like `.sum()`, `.mean()`, etc.) over a fixed-size window that "rolls" through the data. It's commonly used for smoothing trends in time series data. For example, `.rolling(3).sum()` gives the sum of the current and previous 2 rows (3 total). By default, it starts returning results only after the window has enough data. Be careful: always apply `.rolling()` **on the filtered data** if you're working with subsets — otherwise, the trend will be calculated using unwanted rows from the original DataFrame.

-- From ChatGPT


```python
coffee_lat = coffee[coffee["Coffee Type"]=="Latte"].copy()
coffee_lat["trend"] = coffee_lat["Revenue"].rolling(3).sum()
coffee_lat
```
